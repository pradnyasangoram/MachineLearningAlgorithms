```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
x=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
x,y
```

```python
# Fitting Decision tree regression to the dataset
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
y_pred=reg.predict(6.5)
y_pred
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,reg.predict(x),c='b')
plt.title('Truth or Buff(Decision Tree)')
plt.xlabel('Position')
plt.ylabel('Salary')
```

```python
from sklearn.ensemble import RandomForestRegressor
reg1=RandomForestRegressor(n_estimators=10,random_state=0) #no.of tree =10
reg1.fit(x,y)
```

```python
y_pred=reg1.predict(6.5)
y_pred
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,reg1.predict(x),c='b')
plt.title('Truth or Buff(Random Forest)')
plt.xlabel('Position')
plt.ylabel('Salary')
```

```python
plt.subplot(1,3,1)
plt.scatter(x,y,c='r')
plt.plot(x,reg.predict(x),c='b')
plt.title('Truth or Buff(Decision Tree)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.subplot(1,3,3)
plt.scatter(x,y,c='r')
plt.plot(x,reg1.predict(x),c='b')
plt.title('Truth or Buff(Random Forest)')
plt.xlabel('Position')
plt.ylabel('Salary')
```
