```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
ds=pd.read_csv('Position_Salaries-3.csv')
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
x=ds.iloc[:,[0]].values
y=ds.iloc[:,[1]].values
```

```python
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
```

```python
x_train,x_test,y_train,y_test
```

```python
sc_x=StandardScaler()
sc_y=StandardScaler()
sx=sc_x.fit_transform(x_train)
sy=sc_y.fit_transform(y_train)
```

```python
reg=SVR(kernel='poly')
reg.fit(sx,sy)
```

```python
plt.scatter(sx,sy,c='r')
plt.plot(sx,reg.predict(sx),c='b')
plt.title('Salary vs Position')
plt.xlabel('Level')
plt.ylabel('Salary')
```

```python
p=pd.read_csv('Data/Position_Salaries -2 - Pred.csv')
```

```python
p
```

```python
y_pred=p.iloc[:,[0]].values
```

```python
y_pred=reg.predict(sc_x.transform(y_pred))
y_pred=sc_y.inverse_transform(y_pred)
```

```python
y_pred
```

```python
out=pd.concat([p,pd.DataFrame(y_pred,columns=['Salary_pred'])],axis=1)
```

```python
out
```

```python
out.to_csv('Data/Position_Salaries-3_out.csv',index=False)
```
